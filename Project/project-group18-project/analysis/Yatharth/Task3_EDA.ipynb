```python
## Starting with Months April to May, non-graphical analysis.
## we first see the number of total cases from april to may, then we check for may to june, and wehen we
## find the difference we shall be able to compare both 

# April to May
ApriltoMay = df[(df['Date']>'2020-04-18') & (df['Date']<='2020-05-18')]
ApriltoMay.sort_values("Total_Cases",axis=0,ascending=False,inplace=True)

ApriltoMay.head()




```

```python
# May to June
MaytoJune = df[(df['Date']>'2020-05-19') & (df['Date']<='2020-06-18')]
MaytoJune.sort_values("Total_Cases",axis=0,ascending=False,inplace=True)
MaytoJune.head()

```

```python
# July to Spetember

# July to August 
JulytoAugust = df[(df['Date']>'2020-07-18') & (df['Date']<='2020-08-18')]
JulytoAugust.sort_values("Total_Cases",axis=0,ascending=False,inplace=True)
JulytoAugust
```

```python
# August to September
AugusttoSeptember = df[(df['Date']>'2020-08-19') & (df['Date']<='2020-09-18')]
AugusttoSeptember.sort_values("Total_Cases",axis=0,ascending=False,inplace=True)
AugusttoSeptember
```

```python
# Now we compare the total number of cases by subtarcting the total number of cases beteween the month periods and see
# which time period had the most significant rise in case
m1 = ApriltoMay.sum(axis=0,skipna=True)
m2 = MaytoJune.sum(axis=0,skipna=True)
m3 = JulytoAugust.sum(axis=0,skipna=True)
m4 = AugusttoSeptember.sum(axis=0,skipna=True)
p1 = m2["Total_Cases"] - m1["Total_Cases"]
p2 =  m4["Total_Cases"] - m3["Total_Cases"]
print(f" Difference in Cases from time period April to June: {p1}")
print(f" Difference in Cases from time period July to August:{p2}")
```
